```python
import time
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse, parse_qs, urlencode

BASE  = "https://stardewvalleywiki.com"
START = f"{BASE}/Special:AllPages"

HEADERS = {
    # A realistic browser UA avoids some WAF blocks
    "User-Agent": ("Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 "
                   "(KHTML, like Gecko) Chrome/124.0 Safari/537.36"),
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.9",
    "Connection": "keep-alive",
    "Referer": BASE + "/",
    "Upgrade-Insecure-Requests": "1",
}

def normalize_allpages_url(href: str | None) -> str | None:
    """Convert any /mediawiki/index.php?title=Special:AllPages&from=...
    into the pretty /Special:AllPages?from=... form. Preserve other params."""
    if not href:
        return None
    full = urljoin(BASE, href)
    p = urlparse(full)
    qs = parse_qs(p.query)
    # If it's the index.php style with title=Special:AllPages → rebuild as pretty
    if p.path.endswith("/mediawiki/index.php") and qs.get("title", [""])[0] == "Special:AllPages":
        # Keep all query params except 'title'
        params = {k: v[0] for k, v in qs.items() if k != "title"}
        return f"{BASE}/Special:AllPages" + (f"?{urlencode(params)}" if params else "")
    # Already pretty or something else — just return absolute URL
    if p.path.startswith("/Special:AllPages"):
        return full
    return full

def iter_all_pages(max_index_pages: int | None = None):
    s = requests.Session()
    s.headers.update(HEADERS)

    url = START
    seen_index_pages = 0

    while url:
        print("Index:", url)
        r = s.get(url, timeout=30)
        # If they gave us an index.php form that gets blocked, retry normalized
        if r.status_code == 403:
            alt = normalize_allpages_url(url)
            if alt and alt != url:
                url = alt
                r = s.get(url, timeout=30)
        r.raise_for_status()

        soup = BeautifulSoup(r.text, "html.parser")

        # Yield article links from the chunk list
        for a in soup.select(".mw-allpages-chunk li a"):
            yield urljoin(BASE, a["href"])

        # Find the “Next page (...)” anchor inside the nav container(s).
        next_href = None
        # Prefer the LAST nav block (bottom of page)
        for nav in soup.select("div.mw-allpages-nav")[::-1]:
            for a in nav.select("a"):
                if "next page" in a.get_text(strip=True).lower():
                    next_href = a.get("href")
                    break
            if next_href:
                break

        url = normalize_allpages_url(next_href)

        seen_index_pages += 1
        if max_index_pages and seen_index_pages >= max_index_pages:
            break

        # Be polite; increase if you see throttling
        time.sleep(2)
```

```python
import json
import re
import time
from pathlib import Path
from urllib.parse import urljoin, urlparse
import requests
from bs4 import BeautifulSoup

BASE = "https://stardewvalleywiki.com"

HEADERS = {
    # Realistic headers to avoid simple blocks
    "User-Agent": ("Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 "
                   "KHTML, like Gecko) Chrome/124.0 Safari/537.36"),
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.9",
    "Connection": "keep-alive",
    "Referer": BASE + "/",
    "Upgrade-Insecure-Requests": "1",
}

# -----------------------------
# Helpers
# -----------------------------

def fetch_html(url: str) -> str:
    s = requests.Session()
    s.headers.update(HEADERS)
    r = s.get(url, timeout=30)
    # Retry once if rate-limited or transient error
    if r.status_code in (403, 429, 500, 502, 503):
        time.sleep(2.0)
        r = s.get(url, timeout=30)
    r.raise_for_status()
    return r.text

def clean_text(el) -> str:
    """Visible text (collapse whitespace)."""
    return " ".join(el.get_text(" ", strip=True).split())

def resolve(url_or_path: str) -> str:
    return urljoin(BASE, url_or_path)

def extract_links(el):
    links = []
    for a in el.select("a[href]"):
        href = resolve(a["href"])
        txt = a.get_text(" ", strip=True)
        links.append({"text": txt, "href": href})
    return links

def extract_images(el):
    imgs = []
    for img in el.select("img"):
        src = resolve(img.get("src", ""))
        alt = img.get("alt", "")
        w = img.get("width"); h = img.get("height")
        imgs.append({
            "alt": alt,
            "src": src,
            "width": int(w) if (w and str(w).isdigit()) else None,
            "height": int(h) if (h and str(h).isdigit()) else None
        })
    return imgs

def normalize_headers(th_row):
    headers = []
    for th in th_row.select("th"):
        hdr = clean_text(th)
        headers.append(hdr if hdr else "Column")
    # Deduplicate
    seen = {}
    norm = []
    for h in headers:
        c = seen.get(h, 0)
        seen[h] = c + 1
        norm.append(h if c == 0 else f"{h}_{c+1}")
    return norm

def parse_table(table, source_url: str, table_index: int):
    """Return a structured table dict with caption, headers, and rows (text+links+images)."""
    # caption
    caption_el = table.find("caption")
    caption = clean_text(caption_el) if caption_el else ""

    # first row with THs as header row (simple, works for most wikitables)
    header_row = None
    for tr in table.select("tr"):
        if tr.find("th"):
            header_row = tr
            break
    headers = normalize_headers(header_row) if header_row else []

    rows = []
    started = False if header_row else True
    for tr in table.select("tr"):
        if (not started) and tr is header_row:
            started = True
            continue
        if not started:
            continue

        cells = []
        tds = tr.find_all(["td", "th"])
        if not tds:
            continue

        for td in tds:
            cell_text = clean_text(td)
            cell_links = extract_links(td)
            cell_imgs  = extract_images(td)
            cells.append({"text": cell_text, "links": cell_links, "images": cell_imgs})

        if not headers:
            headers = [f"Col{i+1}" for i in range(len(cells))]

        # Align columns
        if len(cells) < len(headers):
            cells += [{"text": "", "links": [], "images": []}] * (len(headers) - len(cells))
        elif len(cells) > len(headers):
            cells = cells[:len(headers)]

        rows.append({h: c for h, c in zip(headers, cells)})

    return {
        "source_url": source_url,
        "table_index": table_index,
        "caption": caption,
        "headers": headers,
        "rows": rows
    }

def extract_tables_from_container(container, source_url: str):
    """Extract all .wikitable tables from a given container (section DOM)."""
    tables = []
    for i, tbl in enumerate(container.select("table.wikitable")):
        tables.append(parse_table(tbl, source_url, i))
    return tables

def sanitize_filename(title: str) -> str:
    s = re.sub(r"[^\w\-_. ]+", "_", title).strip()
    return re.sub(r"\s+", "_", s) or "page"

# -----------------------------
# Core: sectionization (main + H2 sections)
# -----------------------------

def split_into_sections(content_root) -> list:
    """
    Walk the children of #mw-content-text and split content into:
    - 'main' (before first h2)
    - one entry per h2 (title from the heading text)
    Each section keeps its own container of nodes for local extraction.
    """
    sections = []
    current = {"section_title": "main", "nodes": []}

    for child in content_root.children:
        if getattr(child, "name", None) is None:
            continue  # skip NavigableString / whitespace
        # Start a new section on H2
        if child.name and child.name.lower() == "h2":
            # push previous
            if current["nodes"]:
                sections.append(current)
            # new section
            title = clean_text(child) or "Untitled"
            current = {"section_title": title, "nodes": []}
        else:
            current["nodes"].append(child)

    # push the last section
    if current["nodes"]:
        sections.append(current)

    # If there were no nodes at all (edge-case), return a single empty "main"
    if not sections:
        sections = [{"section_title": "main", "nodes": []}]

    return sections

def extract_text_from_nodes(nodes) -> str:
    """
    Within a section's nodes, collect paragraph & list text.
    We ignore infoboxes/navboxes/edit links/scripts/styles that sometimes slip in.
    """
    parts = []
    for node in nodes:
        # strip noisy descendants in-place
        for sel in [".toc", ".navbox", ".infobox", ".mw-editsection", "script", "style"]:
            for el in node.select(sel):
                el.decompose()

        # gather text-bearing elements
        for el in node.find_all(["p", "li", "pre", "code"], recursive=True):
            txt = clean_text(el)
            if txt:
                parts.append(txt)
        # stand-alone text in node itself (rare but just in case)
        if node.name in ("p", "li"):
            txt = clean_text(node)
            if txt:
                parts.append(txt)

    # collapse multiple newlines
    text = "\n".join(parts)
    text = re.sub(r"\n{3,}", "\n\n", text).strip()
    return text

def extract_tables_from_nodes(nodes, source_url: str):
    """
    Find wikitables that appear inside this section's nodes **only**,
    not the entire page. This keeps tables scoped to their section.
    """
    # Build a temporary container to scope the CSS selection
    tmp = BeautifulSoup("<div></div>", "html.parser").div
    for n in nodes:
        tmp.append(n)
    return extract_tables_from_container(tmp, source_url)

# -----------------------------
# Public API
# -----------------------------

def extraction(url: str, out_path: str | None = None) -> str:
    """
    Extract a Stardew Valley Wiki page into JSON with:
      - title
      - link (URL)
      - sections: [{ section_title, text, tables }, ...]
    Writes JSON to `out_path` (file path) or auto-generates from page title.
    Returns the output file path.
    """
    html = fetch_html(url)
    soup = BeautifulSoup(html, "html.parser")

    # Title (prefer #firstHeading when present)
    title_el = soup.select_one("#firstHeading") or soup.select_one("title")
    title = clean_text(title_el) if title_el else urlparse(url).path.rsplit("/", 1)[-1].replace("_", " ") or "Untitled"

    # Main content root
    content = soup.select_one("#mw-content-text") or soup
    # Pre-clean: drop obvious noise outside sections too
    for sel in [".catlinks", ".mw-editsection", ".navbox", "script", "style"]:
        for el in content.select(sel):
            el.decompose()

    # Split into sections (main + h2)
    sections_dom = split_into_sections(content)

    # Build output
    out = {
        "title": title,
        "link": url,
        "sections": []
    }

    for sec in sections_dom:
        sec_title = sec["section_title"] if sec["section_title"] else "main"
        text = extract_text_from_nodes(sec["nodes"])
        tables = extract_tables_from_nodes(sec["nodes"], url)

        out["sections"].append({
            "section_title": sec_title if sec_title.lower() != "contents" else "main",
            "text": text,
            "tables": tables
        })

    # Decide output file
    if out_path is None:
        fname = sanitize_filename(title) + ".json"
        out_dir = Path("data/json_pages")
        out_dir.mkdir(parents=True, exist_ok=True)
        out_path = str(out_dir / fname)

    with open(out_path, "w", encoding="utf-8") as f:
        json.dump(out, f, ensure_ascii=False, indent=2)

    return out_path

# -----------------------------
# CLI test (optional)
# -----------------------------
# if __name__ == "__main__":
#     test_url = "https://stardewvalleywiki.com/Iron_Dirk"
#     fp = extraction(test_url)  # writes to data/json_pages/<Title>.json
#     print("Saved:", fp)
```

```python
import json
import re
import time
from pathlib import Path
from urllib.parse import urlparse
import requests
from bs4 import BeautifulSoup

BASE = "https://stardewvalleywiki.com"

HEADERS = {
    "User-Agent": ("Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 "
                   "KHTML, like Gecko) Chrome/124.0 Safari/537.36"),
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.9",
    "Connection": "keep-alive",
    "Referer": BASE + "/",
    "Upgrade-Insecure-Requests": "1",
}

BANNED_SECTION_TITLES = {"references", "history"}  # case-insensitive compare


# -----------------------------
# Helpers
# -----------------------------

def fetch_html(url: str) -> str:
    """Fetch page HTML with basic retry on transient codes."""
    s = requests.Session()
    s.headers.update(HEADERS)
    r = s.get(url, timeout=30)
    if r.status_code in (403, 429, 500, 502, 503):
        time.sleep(2.0)
        r = s.get(url, timeout=30)
    r.raise_for_status()
    return r.text

def clean_text(el) -> str:
    """Visible text (collapse whitespace)."""
    return " ".join(el.get_text(" ", strip=True).split())

def sanitize_filename(title: str) -> str:
    s = re.sub(r"[^\w\-_. ]+", "_", title).strip()
    return re.sub(r"\s+", "_", s) or "page"


# -----------------------------
# Sectionization (main + H2)
# -----------------------------

def split_into_sections(content_root) -> list:
    """
    Walk the children of #mw-content-text and split content into:
    - 'main' (before first h2)
    - one entry per h2 (title from the heading text)
    Each section keeps its own list of nodes.
    """
    sections = []
    current = {"section_title": "main", "nodes": []}

    for child in content_root.children:
        name = getattr(child, "name", None)
        if name is None:
            continue  # NavigableString / whitespace

        if name.lower() == "h2":
            # push current if it has any nodes
            if current["nodes"]:
                sections.append(current)

            # extract a clean H2 title (handle span.mw-headline etc.)
            headline = child.select_one(".mw-headline")
            title = clean_text(headline if headline else child) or "Untitled"
            current = {"section_title": title, "nodes": []}
        else:
            current["nodes"].append(child)

    if current["nodes"]:
        sections.append(current)

    # if nothing captured (edge case), return a single empty 'main'
    if not sections:
        sections = [{"section_title": "main", "nodes": []}]

    return sections


def extract_text_from_nodes(nodes) -> str:
    """
    Within a section's nodes, collect paragraph & list text.
    Ignore common noisy elements.
    """
    parts = []
    for node in nodes:
        # remove clutter within this node
        for sel in [".toc", ".navbox", ".infobox", ".mw-editsection", "script", "style"]:
            for el in node.select(sel):
                el.decompose()

        # gather text-bearing descendants
        for el in node.find_all(["p", "li", "pre", "code"], recursive=True):
            txt = clean_text(el)
            if txt:
                parts.append(txt)

        # also handle the node itself if it's a paragraph/list
        if getattr(node, "name", "").lower() in ("p", "li"):
            txt = clean_text(node)
            if txt:
                parts.append(txt)

    text = "\n".join(parts)
    text = re.sub(r"\n{3,}", "\n\n", text).strip()
    return text


# -----------------------------
# Public API
# -----------------------------

def extraction(url: str, out_path: str | None = None) -> str:
    """
    Extract a Stardew Valley Wiki page into JSON with:
      - title
      - link (URL)
      - sections: [{ section_title, text }, ...]
    - No tables
    - No per-section links
    - Skips sections titled 'References' or 'History' (case-insensitive)
    Writes JSON to `out_path` or auto-generates from page title.
    Returns the output file path.
    """
    html = fetch_html(url)
    soup = BeautifulSoup(html, "html.parser")

    # Page title
    title_el = soup.select_one("#firstHeading") or soup.select_one("title")
    title = clean_text(title_el) if title_el else urlparse(url).path.rsplit("/", 1)[-1].replace("_", " ") or "Untitled"

    # Main content root
    content = soup.select_one("#mw-content-text") or soup
    # Pre-clean some obvious page-level clutter
    for sel in [".catlinks", ".mw-editsection", ".navbox", "script", "style"]:
        for el in content.select(sel):
            el.decompose()

    # Split content into sections (main + each H2)
    sections_dom = split_into_sections(content)

    out = {
        "title": title,
        "link": url,
        "sections": []
    }

    for sec in sections_dom:
        sec_title_raw = sec["section_title"] or "main"
        sec_title_norm = sec_title_raw.strip()
        # Skip banned sections
        if sec_title_norm.lower() in BANNED_SECTION_TITLES:
            continue

        text = extract_text_from_nodes(sec["nodes"])
        # If both title and text are empty, skip
        if not (sec_title_norm or text):
            continue

        out["sections"].append({
            "section_title": sec_title_norm if sec_title_norm.lower() != "contents" else "main",
            "text": text
        })

    # Decide output path
    if out_path is None:
        fname = sanitize_filename(title) + ".json"
        out_dir = Path("data/json_pages")
        out_dir.mkdir(parents=True, exist_ok=True)
        out_path = str(out_dir / fname)

    with open(out_path, "w", encoding="utf-8") as f:
        json.dump(out, f, ensure_ascii=False, indent=2)

    return out_path


# -----------------------------
# CLI test (optional)
# -----------------------------
if __name__ == "__main__":
    test_url = "https://stardewvalleywiki.com/Weapons"
    saved = extraction(test_url)  # writes to data/json_pages/<Title>.json
    print("Saved:", saved)
```

```python
import json
import re
import time
from pathlib import Path
from urllib.parse import urlparse
import requests
from bs4 import BeautifulSoup

BASE = "https://stardewvalleywiki.com"

HEADERS = {
    "User-Agent": ("Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 "
                   "KHTML, like Gecko) Chrome/124.0 Safari/537.36"),
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.9",
    "Connection": "keep-alive",
    "Referer": BASE + "/",
    "Upgrade-Insecure-Requests": "1",
}

BANNED_SECTION_TITLES = {"references", "history"}  # H2 titles to skip


# -----------------------------
# Helpers
# -----------------------------

def fetch_html(url: str) -> str:
    """Fetch page HTML with a soft retry on transient codes."""
    s = requests.Session()
    s.headers.update(HEADERS)
    r = s.get(url, timeout=30)
    if r.status_code in (403, 429, 500, 502, 503):
        time.sleep(2.0)
        r = s.get(url, timeout=30)
    r.raise_for_status()
    return r.text

def clean_text(el) -> str:
    """Visible text with collapsed whitespace."""
    return " ".join(el.get_text(" ", strip=True).split())

def sanitize_filename(title: str) -> str:
    s = re.sub(r"[^\w\-_. ]+", "_", title).strip()
    return re.sub(r"\s+", "_", s) or "page"


# -----------------------------
# Section roots & text extraction
# -----------------------------

def get_section_root(soup: BeautifulSoup):
    """Prefer the real content container used by MediaWiki."""
    root = soup.select_one("#mw-content-text .mw-parser-output")
    if not root:
        root = soup.select_one("#mw-content-text")
    return root or soup

def extract_text_from_nodes(nodes) -> str:
    """
    Collect paragraph/list/code text from a node list without mutating original DOM.
    """
    html_fragment = "".join(str(n) for n in nodes)
    tmp = BeautifulSoup(html_fragment, "html.parser")

    # Remove common clutter within this scope
    for sel in [".toc", ".navbox", ".infobox", ".mw-editsection", "script", "style"]:
        for el in tmp.select(sel):
            el.decompose()

    parts = []
    for el in tmp.find_all(["p", "li", "pre", "code"], recursive=True):
        txt = clean_text(el)
        if txt:
            parts.append(txt)

    text = "\n".join(parts)
    text = re.sub(r"\n{3,}", "\n\n", text).strip()
    return text


# -----------------------------
# Split H2, then nest H3 inside each H2
# -----------------------------

def split_h2_sections(content_root):
    """
    Returns a list of dicts:
      - {'section_title': 'main', 'nodes': [...]}
      - {'section_title': <H2 title>, 'nodes': [...]}
    """
    sections = []
    current = {"section_title": "main", "nodes": []}

    for child in content_root.children:
        name = getattr(child, "name", None)
        if not name:
            continue  # skip strings/whitespace

        if name.lower() == "h2":
            # push accumulated nodes (if any) as a finished section
            if current["nodes"]:
                sections.append(current)

            # extract clean H2 title
            headline = child.select_one(".mw-headline")
            title = clean_text(headline if headline else child) or "Untitled"
            current = {"section_title": title, "nodes": []}
        else:
            current["nodes"].append(child)

    if current["nodes"]:
        sections.append(current)

    if not sections:
        sections = [{"section_title": "main", "nodes": []}]

    return sections

def split_h3_subsections(nodes):
    """
    Given the node list that belongs to one H2 section, split into:
      - pre-H3 nodes (for the H2 'text')
      - a list of H3 subsections [{'section_title', 'nodes'}]
    """
    # Work over a temporary soup so we can easily iterate block-level children
    html_fragment = "".join(str(n) for n in nodes)
    tmp = BeautifulSoup(html_fragment, "html.parser")

    h2_text_nodes = []
    subsections = []
    current = {"section_title": None, "nodes": []}
    in_h3 = False

    for child in tmp.children:
        name = getattr(child, "name", None)
        if not name:
            # text node/whitespace → attach to current context
            (current["nodes"] if in_h3 else h2_text_nodes).append(child)
            continue

        if name.lower() == "h3":
            # If we were collecting a previous H3, push it
            if current["section_title"] is not None:
                subsections.append(current)
            # Start a new H3 subsection
            headline = child.select_one(".mw-headline")
            title = clean_text(headline if headline else child) or "Untitled"
            current = {"section_title": title, "nodes": []}
            in_h3 = True
        else:
            # Normal content node → goes either to H2 lead or current H3
            (current["nodes"] if in_h3 else h2_text_nodes).append(child)

    # push last H3 if exists
    if current["section_title"] is not None:
        subsections.append(current)

    return h2_text_nodes, subsections


# -----------------------------
# Public API
# -----------------------------

def extraction(url: str, out_path: str | None = None) -> str:
    """
    Extract a Stardew Valley Wiki page into JSON with:
      - title
      - link (page URL only)
      - sections: [
          {
            "section_title": "main" | <H2>,
            "text": "...",
            "subsections": [
               {"section_title": <H3>, "text": "..."},
               ...
            ]
          },
          ...
        ]
    Skips H2 sections titled 'References' or 'History' (case-insensitive).
    Does NOT extract tables or per-section links.
    """
    html = fetch_html(url)
    soup = BeautifulSoup(html, "html.parser")

    # Title
    title_el = soup.select_one("#firstHeading") or soup.select_one("title")
    title = clean_text(title_el) if title_el else urlparse(url).path.rsplit("/", 1)[-1].replace("_", " ") or "Untitled"

    # Section root and pre-clean (page-level)
    content_root = get_section_root(soup)
    for sel in [".catlinks", ".mw-editsection", ".navbox", "script", "style"]:
        for el in content_root.select(sel):
            el.decompose()

    # First split by H2
    h2_sections = split_h2_sections(content_root)

    out = {
        "title": title,
        "link": url,
        "sections": []
    }

    for sec in h2_sections:
        h2_title = sec["section_title"] or "main"

        # Skip banned H2s
        if h2_title.lower() in BANNED_SECTION_TITLES:
            continue

        # For "main" (pre-H2), there is no H3 nesting—just text
        if h2_title == "main":
            text = extract_text_from_nodes(sec["nodes"])
            if text:
                out["sections"].append({
                    "section_title": "main",
                    "text": text,
                    "subsections": []
                })
            continue

        # For normal H2s: split into lead text + nested H3s
        h2_lead_nodes, h3s = split_h3_subsections(sec["nodes"])
        h2_text = extract_text_from_nodes(h2_lead_nodes)

        entry = {
            "section_title": h2_title,
            "text": h2_text,
            "subsections": []
        }

        for h3 in h3s:
            h3_title = (h3["section_title"] or "").strip() or "Untitled"
            h3_text = extract_text_from_nodes(h3["nodes"])
            if h3_text:
                entry["subsections"].append({
                    "section_title": h3_title,
                    "text": h3_text
                })

        # Skip completely empty H2 entries
        if entry["text"] or entry["subsections"]:
            out["sections"].append(entry)

    # Decide output path
    if out_path is None:
        fname = sanitize_filename(title) + ".json"
        out_dir = Path("data/json_pages")
        out_dir.mkdir(parents=True, exist_ok=True)
        out_path = str(out_dir / fname)

    with open(out_path, "w", encoding="utf-8") as f:
        json.dump(out, f, ensure_ascii=False, indent=2)

    return out_path


# -----------------------------
# CLI test (optional)
# -----------------------------
# if __name__ == "__main__":
#     test_url = "https://stardewvalleywiki.com/Crops"
#     saved = extraction(test_url)  # writes to data/json_pages/<Title>.json
#     print("Saved:", saved)

```

```python
import json
import re
import time
from pathlib import Path
from urllib.parse import urlparse
import requests
from bs4 import BeautifulSoup

BASE = "https://stardewvalleywiki.com"

HEADERS = {
    "User-Agent": ("Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 "
                   "KHTML, like Gecko) Chrome/124.0 Safari/537.36"),
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.9",
    "Connection": "keep-alive",
    "Referer": BASE + "/",
    "Upgrade-Insecure-Requests": "1",
}

BANNED_SECTION_TITLES = {"references", "history"}  # H2s to skip (case-insensitive)


# -----------------------------
# Helpers
# -----------------------------

def fetch_html(url: str) -> str:
    """Fetch page HTML with a soft retry on transient codes."""
    s = requests.Session()
    s.headers.update(HEADERS)
    r = s.get(url, timeout=30)
    if r.status_code in (403, 429, 500, 502, 503):
        time.sleep(2.0)
        r = s.get(url, timeout=30)
    r.raise_for_status()
    return r.text

def clean_text(el) -> str:
    """Visible text with collapsed whitespace."""
    return " ".join(el.get_text(" ", strip=True).split())

def sanitize_filename(title: str) -> str:
    s = re.sub(r"[^\w\-_. ]+", "_", title).strip()
    return re.sub(r"\s+", "_", s) or "page"


# -----------------------------
# Section roots & text extraction
# -----------------------------

def get_section_root(soup: BeautifulSoup):
    """Prefer the real content container used by MediaWiki."""
    root = soup.select_one("#mw-content-text .mw-parser-output")
    if not root:
        root = soup.select_one("#mw-content-text")
    return root or soup

def extract_text_from_html_fragment(html_fragment: str) -> str:
    """
    Collect paragraph/list/code text from an HTML fragment string.
    """
    tmp = BeautifulSoup(html_fragment, "html.parser")

    # Remove common clutter within this scope
    for sel in [".toc", ".navbox", ".infobox", ".mw-editsection", "script", "style"]:
        for el in tmp.select(sel):
            el.decompose()

    parts = []
    for el in tmp.find_all(["p", "li", "pre", "code"], recursive=True):
        txt = clean_text(el)
        if txt:
            parts.append(txt)

    text = "\n".join(parts)
    text = re.sub(r"\n{3,}", "\n\n", text).strip()
    return text


# -----------------------------
# Split H2, then flatten H3 into H2 text
# -----------------------------

def split_h2_sections(content_root):
    """
    Returns a list of dicts:
      - {'section_title': 'main', 'nodes': [...]}
      - {'section_title': <H2 title>, 'nodes': [...]}
    """
    sections = []
    current = {"section_title": "main", "nodes": []}

    for child in content_root.children:
        name = getattr(child, "name", None)
        if not name:
            continue  # skip strings/whitespace

        if name.lower() == "h2":
            if current["nodes"]:
                sections.append(current)
            headline = child.select_one(".mw-headline")
            title = clean_text(headline if headline else child) or "Untitled"
            current = {"section_title": title, "nodes": []}
        else:
            current["nodes"].append(child)

    if current["nodes"]:
        sections.append(current)

    if not sections:
        sections = [{"section_title": "main", "nodes": []}]

    return sections

def flatten_h3_into_text(nodes) -> str:
    """
    Given the node list that belongs to one H2 section, produce a single text:
      H2 lead text
      <H3 title>
      <H3 text>
      <next H3 title>
      <next H3 text>
      ...
    """
    # Work over a temporary soup so we can iterate block-level elements
    html_fragment = "".join(str(n) for n in nodes)
    tmp = BeautifulSoup(html_fragment, "html.parser")

    # Split into H2 lead and H3 chunks
    lead_nodes = []
    h3_chunks = []  # list of (title, [nodes])
    current_h3_title = None
    current_h3_nodes = []

    for child in tmp.children:
        name = getattr(child, "name", None)
        if not name:
            # text node/whitespace → attach to current context
            if current_h3_title is None:
                lead_nodes.append(child)
            else:
                current_h3_nodes.append(child)
            continue

        if name.lower() == "h3":
            # Finish previous H3 chunk if any
            if current_h3_title is not None:
                h3_chunks.append((current_h3_title, current_h3_nodes))
            # Start new H3
            headline = child.select_one(".mw-headline")
            current_h3_title = clean_text(headline if headline else child) or "Untitled"
            current_h3_nodes = []
        else:
            if current_h3_title is None:
                lead_nodes.append(child)
            else:
                current_h3_nodes.append(child)

    # Push the last H3 if exists
    if current_h3_title is not None:
        h3_chunks.append((current_h3_title, current_h3_nodes))

    # Build final flattened text
    parts = []

    # H2 lead text
    lead_html = "".join(str(n) for n in lead_nodes)
    lead_text = extract_text_from_html_fragment(lead_html)
    if lead_text:
        parts.append(lead_text)

    # Each H3 title + text
    for title, nodes_list in h3_chunks:
        h3_html = "".join(str(n) for n in nodes_list)
        h3_text = extract_text_from_html_fragment(h3_html)
        if not h3_text:
            continue
        # Add title on its own line to preserve some structure
        parts.append(title)
        parts.append(h3_text)

    text = "\n\n".join(p for p in parts if p.strip())
    return text.strip()


# -----------------------------
# Public API
# -----------------------------

def extraction(url: str, out_path: str | None = None) -> str:
    """
    Extract a Stardew Valley Wiki page into JSON with:
      - title
      - link (page URL only)
      - sections: [{ "section_title": "main"|<H2>, "text": "<flattened text>" }, ...]
    - H3 content is flattened into the parent H2 text with H3 titles inline.
    - Skips H2 sections titled 'References' or 'History' (case-insensitive).
    - Does NOT extract tables or per-section links.
    """
    html = fetch_html(url)
    soup = BeautifulSoup(html, "html.parser")

    # Title
    title_el = soup.select_one("#firstHeading") or soup.select_one("title")
    title = clean_text(title_el) if title_el else urlparse(url).path.rsplit("/", 1)[-1].replace("_", " ") or "Untitled"

    # Section root and pre-clean (page-level)
    content_root = get_section_root(soup)
    for sel in [".catlinks", ".mw-editsection", ".navbox", "script", "style"]:
        for el in content_root.select(sel):
            el.decompose()

    # First split by H2
    h2_sections = split_h2_sections(content_root)

    out = {
        "title": title,
        "link": url,
        "sections": []
    }

    for sec in h2_sections:
        h2_title = sec["section_title"] or "main"

        # Skip banned H2s
        if h2_title.lower() in BANNED_SECTION_TITLES:
            continue

        if h2_title == "main":
            # Everything before first H2 (no H3 here)
            html_fragment = "".join(str(n) for n in sec["nodes"])
            text = extract_text_from_html_fragment(html_fragment)
        else:
            # Flatten H3s into this H2 text
            text = flatten_h3_into_text(sec["nodes"])

        if text:
            out["sections"].append({
                "section_title": "main" if h2_title.lower() == "contents" else h2_title,
                "text": text
            })

    # Decide output path
    if out_path is None:
        fname = sanitize_filename(title) + ".json"
        out_dir = Path("data/json_pages")
        out_dir.mkdir(parents=True, exist_ok=True)
        out_path = str(out_dir / fname)

    with open(out_path, "w", encoding="utf-8") as f:
        json.dump(out, f, ensure_ascii=False, indent=2)

    return out_path


# -----------------------------
# CLI test (optional)
# -----------------------------
# if __name__ == "__main__":
#     test_url = "https://stardewvalleywiki.com/Crops"
#     saved = extraction(test_url)  # writes to data/json_pages/<Title>.json
#     print("Saved:", saved)
```

```python
# save as: build_single_json.py
# usage:   python build_single_json.py

import json
import os
from pathlib import Path

# from stardew_extract import extraction  # <-- import your function

INPUT_LINKS = "links_clean.txt"
OUTPUT_JSON = "data/stardew_all.json"
PER_PAGE_DIR = Path("data/json_pages")  # where extraction() writes the per-page files

def read_links(path: str) -> list[str]:
    links = []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line or line.startswith("#"):
                continue
            links.append(line)
    return links

def main():
    # Ensure output directories exist
    PER_PAGE_DIR.mkdir(parents=True, exist_ok=True)
    Path(OUTPUT_JSON).parent.mkdir(parents=True, exist_ok=True)

    links = read_links(INPUT_LINKS)
    print(f"Found {len(links)} links in {INPUT_LINKS}")

    pages = []
    for i, url in enumerate(links[0:10], 1):
        try:
            print(f"[{i}/{len(links)}] Extracting: {url}")
            out_path = extraction(url)  # returns the path to the per-page JSON
            # Load the page JSON and append to the combined list
            with open(out_path, "r", encoding="utf-8") as f:
                page_obj = json.load(f)
            pages.append(page_obj)
        except Exception as e:
            # Keep going on errors; record a minimal stub for debugging
            print(f"  !! Error on {url}: {e}")
            pages.append({"title": None, "link": url, "sections": [], "error": str(e)})

    # Write the single combined JSON
    combined = {"pages": pages}
    with open(OUTPUT_JSON, "w", encoding="utf-8") as f:
        json.dump(combined, f, ensure_ascii=False, indent=2)

    print(f"\nWrote combined JSON with {len(pages)} pages to: {OUTPUT_JSON}")

if __name__ == "__main__":
    main()
```
